```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r'..\data\raw\data_frame.csv')
df2 = pd.read_csv(r'..\data\raw\data_frame vadim only.csv')
```

```python
df.duplicated().value_counts()
```

```python
df.drop_duplicates(inplace=True)
df2.drop_duplicates(inplace=True)
```

```python
len(df['search_keywords'].unique())
```

```python
fig, ax = plt.subplots()
ax.hist(x=df['search_keywords'], bins=np.arange(-0.5, 36), rwidth=0.7)
ax.set_xticks(ticks=np.arange(0, 36))
plt.xticks(rotation=90)
plt.show()
```
